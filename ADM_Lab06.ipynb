```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/university_student.csv')
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
print(df.fillna)
```

```python
df = pd.get_dummies(df, columns=['Gender', 'School Type'], drop_first=True)
print(df)
```

```python
label_encoder = LabelEncoder()
df['Grade'] = label_encoder.fit_transform(df['Grade'])
```

```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df['Age'] = (pd.to_datetime('today') - df['Date of Birth']).dt.days // 365
df['Total Score'] = df[['Math Score', 'Reading Score', 'Writing Score']].sum(axis=1)
print(df.head())
```

```python
grouped_data = df.groupby('Grade').agg({'Total Score': 'mean', 'Study Time': 'mean', 'Absences': 'sum'})
print(grouped_data)
```

```python
s = df.select_dtypes(include=[np.number])
```

```python
mean_values = s.mean()
print(s.mean)
```

```python
median_values = s.median()
print(s.median)
```

```python
mode_values = s.mode().iloc[0]
print(s.mode)
```
